```python
import pandas as pd
```

```python
# read the cleaned dataset
df = pd.read_csv('census_cleaned.csv')
df.head()
```

```python
# run basic EDA on the dataset using histograms
df.hist(figsize=(10, 10))
```

```python
# run basic EDA on the categorical columns using bar plots
import matplotlib.pyplot as plt

fig, axes = plt.subplots(4, 2, figsize=(15, 20))

df['workclass'].value_counts().plot(kind='bar', ax=axes[0, 0], title='Workclass')
df['education'].value_counts().plot(kind='bar', ax=axes[0, 1], title='Education')
df['marital-status'].value_counts().plot(kind='bar', ax=axes[1, 0], title='Marital Status')
df['occupation'].value_counts().plot(kind='bar', ax=axes[1, 1], title='Occupation')
df['relationship'].value_counts().plot(kind='bar', ax=axes[2, 0], title='Relationship')
df['race'].value_counts().plot(kind='bar', ax=axes[2, 1], title='Race')
df['sex'].value_counts().plot(kind='bar', ax=axes[3, 0], title='Sex')

# Hide the empty subplot
fig.delaxes(axes[3, 1])

plt.tight_layout()
plt.show()
```
